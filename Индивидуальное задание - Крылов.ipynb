## 2.3. Фильтрация данных

```python
filtr1 = df.query('country_name == "China" and 1960 > 1')[['country_name', '1960', '1961', '1962', '1963', '1964', '1965', '2022']]
filtr1 # Фильтрация с использованием метода ".query".
```

```python
filtr2 = df.where((df['country_name'] == "Russian Federation") & (df['1960'] > 100000000))
filtr2 = filtr2.dropna() #Фильтрация с использованием метода ".where".
filtr2 
```

## 2.4. Сводные таблицы

## 2.5. Поиск закономерностей, варианты решения бизнес-задачи

```python
plt.figure(figsize=(65, 10))
df.groupby('country_name')['1960'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Общая популяция по странам в 1960 году')
plt.xlabel('Страна')
plt.ylabel('Популяция в 1960')
plt.show()
```

# 3.0. Графический анализ данных

## 3.1. Диаграммы, библиотека Matplotlib
Распределение adr_rating дает нам понять, что большинство игроков имеет рейтинг < 1. 

```python
plt.figure(figsize=(10, 6))
plt.hist(df['1960'], bins=15, edgecolor='k', alpha=0.7)
plt.title('Распределение популяции по странам в 1960-ом году')
plt.xlabel('Популяция')
plt.ylabel('Количество стран')
plt.grid(True)
plt.show()
```

## 3.2. Диаграммы, библиотека Seaborn

```python
plt.figure(figsize=(40, 80))
sea.scatterplot(x='1960', y='1961', data=df, hue='country_name', palette='viridis')
plt.title('Зависимость популяции от времени(переход с 1960 - 1961 года')
plt.xlabel('Популяция 1960')
plt.ylabel('Популяция 1961')
plt.legend(title='Страна')
plt.grid(True)
plt.show()
```
